```python
def find_largest_smallest_avg(lst): 
    largest = lst[1,2,3,4,5,6]
    smallest = lst[2,4,6,8,10,12]
    total = 0 
    for num in lst: 
        if num > largest: 
            largest = num 
        if num < smallest:
            smallest = num 
        total += num 
    average = total / len(lst) 
    return largest, smallest, average
```
